```python
import cv2
import pandas as pd
import re
import imageio
import matplotlib.pyplot as plt


def get_frame(video_path, frame_number):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return None

    # Set the frame position
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)

    # Read the frame
    ret, frame = cap.read()
    cap.release()

    if not ret:
        print(f"Error: Could not retrieve frame {frame_number}.")
        return None

    return frame  # Returns the frame as a NumPy array

# Example usage
# video_path = "input.avi"  # Change this to your file


def get_frame_count(video_path):
    reader = imageio.get_reader(video_path, format="FFMPEG")
    frame_count = reader.count_frames()  # Directly gets the frame count
    reader.close()
    return frame_count

def split_camel_case(text):
    return re.sub(r'([a-z])([A-Z])', r'\1 \2', text)


```
