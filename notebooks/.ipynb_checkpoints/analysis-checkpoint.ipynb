```python
from PIL import Image
import numpy as np

pivot = 200
img_grid = np.zeros((512*6, 512* 10, 3))
for i in range(0, 20):
    
    row = (i //10) * 3
    col = i % 10
    img_prior = np.array(Image.open(meta_data.loc[pivot + i].hr_prior).convert("RGB"))/255
    lr = np.array(Image.open(meta_data.loc[pivot + i].lr_paths).convert("RGB"))/255
    img= np.array(Image.open(meta_data.loc[pivot + i].hr_recon).convert("RGB"))/255
    
    img_grid[row * 512 : (row+1) * 512, col * 512: (col+1)*512, :] = img_prior
    img_grid[(row+1) * 512 : (row+2) * 512, col * 512: (col+1)*512, :] = lr
    img_grid[(row+2) * 512 : (row+3) * 512, col * 512: (col+1)*512, :] = img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20, 10))
plt.imshow(img_grid)
```

```python
img_grid = np.zeros((512*4, 512* 10, 3))
for i in range(20):
    
    row = (i //10) * 2
    col = i % 10
    img_prior = np.array(Image.open(meta_data.loc[i].lr_paths).convert("RGB"))/255
    img= np.array(Image.open(meta_data.loc[i].hr_recon).convert("RGB"))/255
    
    img_grid[row * 512 : (row+1) * 512, col * 512: (col+1)*512, :] = img_prior
    img_grid[(row+1) * 512 : (row+2) * 512, col * 512: (col+1)*512, :] = img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20, 10))
plt.imshow(img_grid)
```
